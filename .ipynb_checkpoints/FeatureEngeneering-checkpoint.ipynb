```python
# imports
import pandas as pd
from dateutil import parser

```

```python
# read dataframe
def read_df():
    df = pd.read_csv('../Datasets/alarms.csv', sep=';')
    df[['start', 'end']] = df[['start', 'end']].apply(pd.to_datetime)
    df.sort_values(by=['start'])
    
    #debug
    #df = df[df['region_id'] == 3]
    
    return df  #.head(30)

def read_df_custom(filename):
    df = pd.read_csv(filename, sep=';')
    df[['start', 'end']] = df[['start', 'end']].apply(pd.to_datetime)
    df.sort_values(by=['start'])
    
    #debug
    #df = df[df['region_id'] == 3]
    
    return df  #.head(30)

def save_df(df, name):
    df.to_csv('../FeatureEngineering/' + name + '.csv', sep=';')
    
```

```python
def alarm_past_24_hours(df):
    df['alarms_past_24'] = 0
    for index, row in df.iterrows():
        count = len(df[(df['region_id'] == row['region_id']) & (df['end'] > row['start'] - pd.Timedelta(hours=24)) & (df['end'] < row['start'])].index)
        df['alarms_past_24'][index] = count
        #print(count)
    
```

```python
df = read_df();
alarm_past_24_hours(df)

save_df(df, 'ver1')
df
```

```python
def count_alarm_overlap(df):
    df['simultaneous_alarms'] = -1
    for index, row in df.iterrows():
        filter1 = (row['start'] < df['start']) & (df['start'] < row['end']) # start other is in the middle of current alert
        filter2 = (row['start'] < df['end']) & (df['end'] < row['end'])     # onther end in the middle of current alert
        filter3 = (df['start'] < row['start']) & (row['start'] < df['end']) # this alert inside of the other one
        
        count = len(df[filter1 | filter2 | filter3].index)
        df['simultaneous_alarms'][index] = count
        #print(count)
```

```python
df = read_df()
print('read')
alarm_past_24_hours(df)
print('got 24')
count_alarm_overlap(df)
print('got simultaneous')
save_df(df, 'ver2')
print('saved')
df
```

```python
def hours_from_prev_alarm(df):
    df['hours_from_last_alarm'] = -1
    for index, row in df.iterrows():
        shortened_df = df.head(index)
        filter1 = (shortened_df['region_id'] == row['region_id'])
        value = pd.Timedelta(row['start'] - shortened_df[filter1]['end'].max()).seconds / 360;
        df['hours_from_last_alarm'][index] = value
```

```python
df = read_df_custom('../FeatureEngineering/ver2.csv');
print(df[['region_id', 'start', 'end']].head(15))
hours_from_prev_alarm(df)
save_df(df, 'ver3')
```

```python
def holiday_is_near(df):
    holiday_df = pd.read_csv('../Datasets/holidays.csv', sep=';')
    print(holiday_df)
    holiday_df['date'] = holiday_df['date'].apply(pd.to_datetime)
    holiday_df = holiday_df.sort_values(by=['date'])
    holiday_df = holiday_df.set_index('date')
    
    df['holiday_is_near'] = False
    #df['holiday_date'] = 0
    
    for index, row in df.iterrows():
        closest_holiday = holiday_df.index[holiday_df.index.get_loc(row['start'], method='nearest')]
        value = abs(pd.Timedelta(row['start'] - closest_holiday).days) <= 3
        #print(value)
        #value = pd.Timedelta().seconds / 360;
        df['holiday_is_near'][index] = value
        #df['holiday_date'][index] = closest_holiday
```

```python
df = read_df_custom('../FeatureEngineering/ver3.csv')#.head(5000);
#print(df[['region_id', 'start', 'end']].head(15))
holiday_is_near(df)
save_df(df, 'ver6')
```
