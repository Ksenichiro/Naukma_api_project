```python
# imports
import pandas as pd
from dateutil import parser

```

```python
# read dataframe
def read_df():
    df = pd.read_csv('../FeatureEngineering/giant_take2_ver2.csv', sep=';')
    df[['event_start', 'event_end']] = df[['event_start', 'event_end']].apply(pd.to_datetime)
    
    return df #.head(300)

def read_df_custom(filename):
    df = pd.read_csv(filename, sep=';')    
    #debug
    #df = df[df['region_id'] == 3]
    
    return df  #.head(30)

def save_df(df, name):
    df.to_csv('../FeatureEngineering/' + name + '.csv', sep=';')
    
```

```python
def alarm_past_24_hours(df):
    df_alarms_only = read_df_custom('../Datasets/alarms.csv')
    df['event_alarms_past_24'] = 0
    print('started')
    for index, row in df.iterrows():
        datetime = parser.parse(row['day_datetime'] + ' ' + row['hour_datetime'])
        filter1 = (df_alarms_only['region_id'] == row['region_id']) & (df_alarms_only['end'] > datetime - pd.Timedelta(hours=24)) & (df_alarms_only['end'] < datetime)
        count = len(df_alarms_only[filter1].index)
        df['event_alarms_past_24'][index] = count
        if(index % 1000 == 0):
            print(index)
    
    print('done')
    
```

```python
df = read_df();
alarm_past_24_hours(df)

save_df(df, 'giant_take2_ver1')
df
```

```python
def count_alarm_overlap(df):
    df_alarms_only = read_df_custom('../Datasets/alarms.csv')
    df['event_simultaneous_alarms'] = -1
    print('started')
    for index, row in df.iterrows():
        datetime = parser.parse(row['day_datetime'] + ' ' + row['hour_datetime'])
        filter1 = (df_alarms_only['start'] < datetime) & (df_alarms_only['end'] > datetime)
        #filter1 = (row['start'] < df_alarms_only['start']) & (df_alarms_only['start'] < row['end']) # start other is in the middle of current alert
        #filter2 = (row['start'] < df_alarms_only['end']) & (df_alarms_only['end'] < row['end'])     # onther end in the middle of current alert
        #filter3 = (df_alarms_only['start'] < row['start']) & (row['start'] < df_alarms_only['end']) # this alert inside of the other one
        
        count = len(df_alarms_only[filter1].index)
        df['event_simultaneous_alarms'][index] = count
        if(index % 1000 == 0):
            print(index)
```

```python
df = read_df()
count_alarm_overlap(df)
print('got simultaneous')
save_df(df, 'giant_take2_ver3')
print('saved')
df
```

```python
def hours_from_prev_alarm(df):
    df['event_hours_from_last_alarm'] = -1
    df_alarms_only = read_df_custom('../Datasets/alarms.csv')
    df_alarms_only[['start', 'end']] = df_alarms_only[['start', 'end']].apply(pd.to_datetime)
    
    for index, row in df.iterrows():
        datetime = parser.parse(row['day_datetime'] + ' ' + row['hour_datetime'])
        filter1 = (df_alarms_only['region_id'] == row['region_id']) & (df_alarms_only['end'] < datetime)
        value = pd.Timedelta(datetime - df_alarms_only[filter1]['end'].max()).seconds / 3600;
        df['event_hours_from_last_alarm'][index] = value
        if(index % 1000 == 0):
            print(index)
```

```python
df = read_df_custom('../FeatureEngineering/giant_take2_ver2.csv');
df[['event_start', 'event_end']] = df[['event_start', 'event_end']].apply(pd.to_datetime)

#print(df[['region_id', 'start', 'end']].head(15))
hours_from_prev_alarm(df)
save_df(df, 'giant_take2_ver3')
```

```python
def holiday_is_near(df):
    holiday_df = pd.read_csv('../Datasets/holidays.csv', sep=';')
    print(holiday_df)
    holiday_df['date'] = holiday_df['date'].apply(pd.to_datetime)
    holiday_df = holiday_df.sort_values(by=['date'])
    holiday_df = holiday_df.set_index('date')
    
    df['event_holiday_is_near'] = False
    #df['holiday_date'] = 0
    
    for index, row in df.iterrows():
        datetime = parser.parse(row['day_datetime'])
        closest_holiday = holiday_df.index[holiday_df.index.get_loc(datetime, method='nearest')]
        value = abs(pd.Timedelta(row['start'] - closest_holiday).days) <= 3
        #print(value)
        #value = pd.Timedelta().seconds / 360;
        df['holiday_is_near'][index] = value
        #df['holiday_date'][index] = closest_holiday
```

```python
df = read_df_custom('../FeatureEngineering/ver3.csv')#.head(5000);
#print(df[['region_id', 'start', 'end']].head(15))
holiday_is_near(df)
save_df(df, 'ver6')
```
