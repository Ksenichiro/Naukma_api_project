```python
import datetime
import calendar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
REPORTS_DATA_FOLDER = "data/2_isw_preprocessed"
REPORTS_DATA_FILE = "all_days.csv"

TFIDF_NUMBER = 100

EVENTS_DATA_FOLDER = "data/1_events"
EVENTS_DATA_FILE = "alarms.csv"

PLOTS_DATA_FOLDER = "data/5_plots"

WEATHER_DATA_FOLDER = "data/1_weather"
WEATHER_DATA_FILE = "all_weather_by_hour_v2.csv"

REGIONS_DATA_FOLDER = "data/0_meta"
REGIONS_DATA_FILE = "regions.csv"

MODEL_FOLDER = "model"

OUTPUT_FOLDER = "data/4_all_data_preprocessed"
ISW_OUTPUT_DATA_FILE = "all_isw.csv"
WEATHER_EVENTS_OUTPUT_DATA_FILE = "all_hourly_weather_events_v2.csv"
WEATHER_EVENTS_KEYWORDS_OUTPUT_DATA_FILE = "all_hourly_weather_events_isw_v2.csv"
```

```python
df = pd.read_csv(f"{EVENTS_DATA_FOLDER}/{EVENTS_DATA_FILE}", delimiter=";")

# Iterate over each region and plot the stats per day
for region_id in range(1, 25):
    if region_id == 11:
        continue
    # Filter the data by region
    region_df = df.loc[df["region_id"] == region_id]

    # Convert the "start" column to a datetime object
    region_df["start"] = pd.to_datetime(region_df["start"])

    # Create a new dataframe with the date as the index and count the number of occurrences for each date
    stats_per_day = region_df.groupby(region_df["start"].dt.date).size().reset_index(name="count")
    stats_per_day.set_index("start", inplace=True)

    # Plot the resulting dataframe
    plt.plot(stats_per_day.index, stats_per_day["count"])
    plt.title(region_df.iloc[0]["region_title"])
    plt.xlabel("Date")
    plt.ylabel("Count")
    plt.gcf().autofmt_xdate()
    region = region_df.iloc[0]["region_title"]
    plt.savefig(f"{PLOTS_DATA_FOLDER}/{region}.png")
    plt.show()
    
    
```

```python

# Read the CSV file
df = pd.read_csv(f"{WEATHER_DATA_FOLDER}/{WEATHER_DATA_FILE}")


# Define the type of weather to count
weather_type = "snow"  # Change this to the type of weather you want to count

# Create a date range from 2022-01-01 to 2022-12-31
date_range = pd.date_range(start="2022-01-01", end="2022-12-31", freq="D")


# Loop through each unique city_address
unique_values = df["city_address"].unique()
print(unique_values)

for city in unique_values:
    # Filter the data for the current city and weather type
    #print(df)
    city_df = df[(df["city_address"] == city)]
    #  (df["day_conditions"].str.lower().str.contains(weather_type))
    
    
    df_new = city_df[['day_datetime', 'day_conditions']]
    df_new["day_datetime"] = pd.to_datetime(df_new["day_datetime"])
    weather_hours = {}
    
    date_range = pd.date_range(start="2022-01-01", end="2022-12-31")

    # Loop through each date in the range
    for date in date_range:
        #city_df = df[(df["city_address"] == city) & (df["day_precipprob"] > 0) & (df["day_conditions"].str.lower().str.contains(weather_type))]
        #
        date_frame = df_new[(df_new["day_datetime"] == date) & (df_new["day_conditions"].str.lower().str.contains(weather_type))]
        #rint(date)
        #print(date_frame)
        weather_hours[date] =  len(date_frame)
    #print(weather_hours)

    # Convert the dictionary to a dataframe
    df_to_plot = pd.DataFrame(list(weather_hours.items()), columns=['Date', 'Hours'])

    # Convert the date column to a datetime object
    df_to_plot['Date'] = pd.to_datetime(df_to_plot['Date'])

    # Set the date column as the index
    df_to_plot = df_to_plot.set_index('Date')

    # Plot the data
    df_to_plot.plot(kind='line', x_compat=True)

    # Add labels
    plt.xlabel('Date')
    plt.ylabel('Hours')
    plt.title(city)
    
    plt.savefig(f"{PLOTS_DATA_FOLDER}/snow/{city}.png")
    # Show the plot
    plt.show()

```

```python
# Read the CSV file
df = pd.read_csv(f"{WEATHER_DATA_FOLDER}/{WEATHER_DATA_FILE}")


# Create a date range from 2022-01-01 to 2022-12-31
date_range = pd.date_range(start="2022-01-01", end="2022-12-31", freq="D")


# Loop through each unique city_address
unique_values = df["city_address"].unique()
print(unique_values)

for city in unique_values:
    df = pd.read_csv(f"{WEATHER_DATA_FOLDER}/{WEATHER_DATA_FILE}")
    # Filter the data for the current city and weather type
    #print(df)
    city_df = df[(df["city_address"] == city)]
    #  (df["day_conditions"].str.lower().str.contains(weather_type))
    
    
    df_new = city_df[['day_datetime', 'day_pressure']]
    #print(df_new)
    df_new["day_datetime"] = pd.to_datetime(df_new["day_datetime"])
    mid_preasure = {}
    
    date_range = pd.date_range(start="2022-01-01", end="2022-12-31")

    # Loop through each date in the range
    for date in date_range:
        #city_df = df[(df["city_address"] == city) & (df["day_precipprob"] > 0) & (df["day_conditions"].str.lower().str.contains(weather_type))]
        #
        date_frame = df_new[(df_new["day_datetime"] == date)]
        preasure = date_frame['day_pressure'].mean()
        #print(preasure)
        #print(date_frame)
        mid_preasure[date] =  preasure
    #print(weather_hours)

    # Convert the dictionary to a dataframe
    df_to_plot = pd.DataFrame(list(mid_preasure.items()), columns=['Date', 'preasure'])

    # Convert the date column to a datetime object
    df_to_plot['Date'] = pd.to_datetime(df_to_plot['Date'])

    # Set the date column as the index
    df_to_plot = df_to_plot.set_index('Date')

    # Plot the data
    df_to_plot.plot(kind='line', x_compat=True)

    # Add labels
    plt.xlabel('Date')
    plt.ylabel('preasure')
    plt.title(city)
    
    plt.savefig(f"{PLOTS_DATA_FOLDER}/preasure/{city}.png")
    # Show the plot
    plt.show()
```
